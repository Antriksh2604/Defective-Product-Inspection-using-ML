```python
resize_rescale_layers = tf.keras.Sequential([
    Resizing(CONFIGURATION['IM_SIZE'],CONFIGURATION['IM_SIZE']),
    Rescaling(1.0/225),
])
```

```python
cnn_model = tf.keras.Sequential([
    InputLayer(input_shape=(None,None,3)),

    resize_rescale_layers,
    
    Conv2D(filters=CONFIGURATION['N_FILTERS'], kernel_size= CONFIGURATION['KERNEL_SIZE'], strides= CONFIGURATION['N_STRIDES'], padding= 'valid', activation='relu'),
    BatchNormalization(),
    MaxPool2D(pool_size= CONFIGURATION['POOL_SIZE'], strides=2*CONFIGURATION['N_STRIDES']),
    
    Conv2D(filters=3*CONFIGURATION['N_FILTERS']-2,kernel_size= CONFIGURATION['KERNEL_SIZE'], strides= CONFIGURATION['N_STRIDES'], padding= 'valid', activation='relu'),
    BatchNormalization(),
    MaxPool2D(pool_size= CONFIGURATION['POOL_SIZE'] , strides=2*CONFIGURATION['N_STRIDES']),
    
    Flatten(),
     
    Dense(CONFIGURATION['N_DENSE_1'], activation='relu'),
    BatchNormalization(),
    
    Dense(CONFIGURATION['N_DENSE_2'], activation='relu'),
    BatchNormalization(),
    
    Dense(CONFIGURATION['NUM_CLASSES'], activation='sigmoid')
])
cnn_model.summary()
```

```python
cnn_model.trainable_weights
```

```python
trainable_params = sum(count_params(layer) for layer in cnn_model.trainable_weights)
non_trainable_params = sum(count_params(layer) for layer in cnn_model.non_trainable_weights)
p1 = trainable_params + non_trainable_params
```

EfficientNet with Learnable Resizer

```python
INP_SIZE = (300, 300)
TARGET_SIZE = (256, 256)
INTERPOLATION = "gaussian"
```

```python
layers = tf.keras.layers
def conv_block(x, filters, kernel_size, strides, activation=layers.LeakyReLU(0.2)):
    x = layers.Conv2D(filters, kernel_size, strides, padding="same", use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    if activation:
        x = activation(x)
    return x


def res_block(x):
    inputs = x
    x = conv_block(x, 16, 3, 1)
    x = conv_block(x, 16, 3, 1, activation=None)
    return layers.Add()([inputs, x])

    # Note: user can change num_res_blocks to >1 also if needed


def get_learnable_resizer(filters=16, num_res_blocks=1, interpolation=INTERPOLATION):
    inputs = layers.Input(shape=[None, None, 3])

    # First, perform naive resizing.
    naive_resize = layers.Resizing(*TARGET_SIZE, interpolation=interpolation)(inputs)

    # First convolution block without batch normalization.
    x = layers.Conv2D(filters=filters, kernel_size=7, strides=1, padding="same")(inputs)
    x = layers.LeakyReLU(0.2)(x)

    # Second convolution block with batch normalization.
    x = layers.Conv2D(filters=filters, kernel_size=1, strides=1, padding="same")(x)
    x = layers.LeakyReLU(0.2)(x)
    x = layers.BatchNormalization()(x)

    # Intermediate resizing as a bottleneck.
    bottleneck = layers.Resizing(*TARGET_SIZE, interpolation=interpolation)(x)

    # Residual passes.
    # First res_block will get bottleneck output as input
    x = res_block(bottleneck)
    # Remaining res_blocks will get previous res_block output as input
    for _ in range(num_res_blocks - 1):
        x = res_block(x)

    # Projection.
    x = layers.Conv2D(
        filters=filters, kernel_size=3, strides=1, padding="same", use_bias=False
    )(x)
    x = layers.BatchNormalization()(x)

    # Skip connection.
    x = layers.Add()([bottleneck, x])

    # Final resized image.
    x = layers.Conv2D(filters=3, kernel_size=7, strides=1, padding="same")(x)
    final_resize = layers.Add()([naive_resize, x])

    return tf.keras.Model(inputs, final_resize, name="learnable_resizer")


learnable_resizer = get_learnable_resizer()
```

```python
learnable_resizer.summary()
```

```python
def get_model():

    inputs = Input((None, None, 3))
    x = Rescaling(scale=1.0 / 255)(inputs)
    x = learnable_resizer(x)
    x = EfficientNet_backbone(x)
    x = GlobalAveragePooling2D()(x)
    x = Dense( CONFIGURATION['N_DENSE_1'], activation= 'relu')(x)
    x = BatchNormalization()(x)
    x = Dense( CONFIGURATION['N_DENSE_2'], activation= 'relu')(x)
    outputs = Dense( CONFIGURATION['NUM_CLASSES'], activation= 'sigmoid')(x)

    return tf.keras.Model(inputs, outputs)
```

```python
def train(config=None):
    with wandb.init(entity='antrikshraj',project='Defect Detection',config=CONFIGURATION) as run:
        config = wandb.config
        artifact = run.use_artifact('antrikshraj/Defect Detection/train_dataset:v0', type='raw_data')
        artifact_dir = artifact.download()
        
        artifact = run.use_artifact('antrikshraj/Defect Detection/val_dataset:v0', type='raw_data')
        artifact_dir = artifact.download()
        

        trained_model = wandb.Artifact(
            name = "Trained_simple_cnn_model",
            type = "model",
            description= "A trained version of our model",
            metadata = CONFIGURATION
            
        )
        train_artifact_dir = "./artifacts/train_dataset-v0"
        train_dataset = tf.data.Dataset.load(path=train_artifact_dir)
        val_artifact_dir = "./artifacts/val_dataset-v0"
        combined_test_dataset = tf.data.Dataset.load(path=val_artifact_dir)

        DATASET_SIZE = len(combined_test_dataset)
        val_size = int(0.8 * DATASET_SIZE)
        
        val_dataset = combined_test_dataset.take(val_size)
        training_dataset = (
            train_dataset
            .batch(config.BATCH_SIZE,drop_remainder=True)
            .prefetch(tf.data.AUTOTUNE)
        )
        validation_dataset = (
            val_dataset
            .batch(config.BATCH_SIZE,drop_remainder=True)
            .prefetch(tf.data.AUTOTUNE)
        )
        artifact = run.use_artifact('antrikshraj/Defect Detection/Untrained_simple_cnn_model:v0', type='model')
        artifact_dir = artifact.download()

        model_artifact_file = "./artifacts/Untrained_simple_cnn_model-v0/untrained_simple_cnn_model.h5"
        model = tf.keras.models.load_model(model_artifact_file)

        model.compile(
           optimizer = Adam(learning_rate= config.LEARNING_RATE),
             loss = 'binary_crossentropy',
            metrics = [TruePositives(name= 'tp'), FalsePositives(name='fp'), TrueNegatives(name='tn'), FalseNegatives(name='fn'),BinaryAccuracy(name='accuracy'),Precision(name='precision'),Recall(name='recall'),AUC(name='auc')]

        )
        model.fit(training_dataset,
                  validation_data=validation_dataset,epochs = config.N_EPOCHS, callbacks = [WandbCallback()])

count = 10
# no of runs to execute
wandb.agent(sweep_id,function=train,count=count)
```

```python
def log_confusion_matrix(config=None):
    with wandb.init(entity='antrikshraj',project='Defect Detection',config=CONFIGURATION) as run:
        config = wandb.config
        
        artifact = run.use_artifact('antrikshraj/Defect Detection/model-clear-sweep-5:v4', type='model')
        artifact_dir = artifact.download()
        
        model = tf.keras.models.load_model("./artifacts/model-clear-sweep-5-v4/")

        artifact = run.use_artifact('antrikshraj/Defect Detection/val_dataset:v0', type='raw_data')
        artifact_dir = artifact.download()

        val_artifact_dir = "./artifacts/val_dataset-v0"
        combined_test_dataset = tf.data.Dataset.load(path=val_artifact_dir)
        
        test_dataset = (
            combined_test_dataset
            .batch(config.BATCH_SIZE,drop_remainder=True)
            .prefetch(tf.data.AUTOTUNE)
        )
        
        labels = []
        inp = []

        for x,y in test_dataset.unbatch().as_numpy_iterator():
            labels.append(y)

        labels = np.array([i[0] for i in labels])
        predicted = model.predict(test_dataset)
        #print('labels',labels,labels.dtype)
        #print('predicted',predicted,predicted.dtype)

        pred = []
        for i in range(len(predicted)):
            if (predicted[i][0]<0.5):
                pred.append([1,0])

            else:
                pred.append([0,1])

        pred = np.array(pred)

        cm = wandb.plot.confusion_matrix(
            y_true=labels, probs = pred, class_names= config.CLASS_NAMES
        )
        
        wandb.log({"conf_mat": cm})
        
```
